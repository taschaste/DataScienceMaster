### Problem 1 (2 points)
Write function `mean_by_group(grades, groups)` that takes two series: `grades` contains grades of some students and `groups` their group numbers (in the same order). The function should return dictionary with mean grade of each group.

For example, for `grades = pd.Series([5, 4, 3, 5, 2])` and `groups = pd.Series(["B", "A", "A", "B", "A"])`, function should return dictionary `{'A': 3.0, 'B': 5.0}`.

**Note.** No loops and `if`'s are allowed.

**Hint.** You can create dataframe which columns are `grades` and `groups`, for example, by passing `pd.DataFrame` dictionary whose keys are column names and values are columns. `pd.Series` object has method `to_dict()`.

````python
# YOUR CODE HERE:

def mean_by_group(grades, groups):
    
    grades = pd.Series(grades, name='grades')
    groups = pd.Series(groups, name='groups')

    df = pd.merge(grades, groups, right_index=True, left_index=True)
    res = df.groupby("groups")["grades"].mean()
        
    return res.to_dict()
````

Write function `make_panel(df)` that takes dataframe, which rows are firms and columns are years and cells are some numeric value. Your function should return dataframe with three columns: `year`, `firm` and `value` (in this particular order). Table should be sorted by firms, and firms coincide, then by years. Index should be standard (0, 1, 2, ...).

For example, from table

````python
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]], columns=[2000, 2001], 
             index=['Firm_A', 'Firm_B', 'Firm_C'])
df
````

your function should make and return the following table:

````python
out = pd.DataFrame([[2000, 'Firm_A', 1],
                    [2001, 'Firm_A', 2],
                    [2000, 'Firm_B', 3],
                    [2001, 'Firm_B', 4],
                    [2000, 'Firm_C', 5],
                    [2001, 'Firm_C', 6]], columns=['year', 'firm', 'value'])
out
````

**Hint.** Most probably, you need methods `melt`, `reset_index` and `rename`.

**Note.** No loops and `if`'s are allowed!

````python
# YOUR CODE HERE

def make_panel(df):
    
    # create a new data frame
    df_new = pd.melt(df.reset_index(), id_vars=['index'], value_vars=list(df.columns), var_name='year', value_name='value')
    
    # rename the column
    df_new.rename(columns={'index':'firm'}, inplace=True)
    
    # change order
    df_new = df_new[['year', 'firm', 'value']]
    
    # sort the data
    df_new = df_new.sort_values(['firm', 'year'], ascending=[True, True])
    
    # reset indexes
    df_new.reset_index(drop=True, inplace=True)
    
    return df_new
````

````python
def test():
    import random
    import string
    def test_panel(num_of_firms, num_of_years, seed=10):
        d = {}
        random.seed(seed)
        n = []
        f = []
        for i in range(num_of_firms):
            x = ''.join(random.choices(string.ascii_uppercase, k=3))
            f.extend([x] * num_of_years)
            y = random.sample(range(max(num_of_firms, num_of_years) * 2), num_of_years)
            n.extend([y])
            d[x] = y
        d = pd.DataFrame(d).T
        d.columns = [x for x in range(1990, 1990 + num_of_years)]
        outp = pd.DataFrame({'year': [x for x in range(1990, 1990 + num_of_years)] * num_of_firms, 'firm': f, 'value': sum(n, [])}).sort_values(['firm', 'year']).reset_index(drop=True)
        pd.testing.assert_frame_equal(outp[['year', 'firm', 'value']], make_panel(d), check_dtype=False)
    test_panel(1, 1)
    test_panel(10, 20)
    test_panel(15, 150)
    test_panel(150, 15)
    test_panel(1, 10)
test()
````

**Note.** You *may* use `for` that has not more than two lines inside *or* one `for` and one list comprehension. (In fact, problem is solvable with only one list comprehension allowed.)

**Hint.** Function `pd.concat` can be useful here.

````python
# YOUR CODE HERE

def merge_gradebooks(gradebooks):
    
    data = [gradebooks[i][1] for i in range(len(gradebooks))]
    for i in range(len(gradebooks)):
        data[i]['group'] = gradebooks[i][0]
    

    df = pd.concat(data)
   
    return df.sort_values(by=list(df.columns), ascending=True)
    
````

Assume you are solving Problem 2, but you don't have prices in the table `df`. Instead, you have different dataframe, `prices`, that contains two columns: one with item names and another is its price. For example: 

```
df:
        Customer       Item  Quantity  
    0      Alice       Milk         1
    1        Bob       Milk         2
    2      Alice      Beans        10
    3  Elizabeth  Chocolate         5

prices:
       item_name  price
    0       Milk     10
    1      Beans     20
    2  Chocolate     30
    3       Rice      5
```

Write function `check_table_merge(df, prices)` that do the same thing as `check_table` but uses information in the dataframe `prices`.

**Note.** It is guaranteed that each item that occurs in `df` also occure in `prices` exactly one time, however, it is possible that there are items in `prices` that does not occur in `df`. Decide which type of merge/join you need here.

````python
# YOUR CODE HERE

def check_table_merge(df, prices):
    
    # rename the column
    prices.rename(columns={'item_name':'Item'}, inplace=True)
    
    # merge with prices
    df = df.merge(prices, how='left', on='Item')              

    # add a new column with spent sum
    df['Spent Sum'] = df['Quantity'] * df['price']

    
    res = df.groupby("Customer")["Spent Sum"].sum()
    res2 = df.groupby("Item")["Spent Sum"].sum()
        
    return (res.to_dict(), res2.to_dict())

    
````

### Problem 6 (3 points)
There is a federated country that consists of several lands. Every land can send a particular number of delegates to the convention where president will be chosen. Different lands have different number of delegates. Every land conduct elections on its territory, then instructs its delegates to vote as the majority of voters decide (the winner takes all). President is elected by the majority of delegate's votes.

You have to write function `winner_votes(results)`, that returns a two-tuple: first element is the winner name and second is the number of delegates who votes for him. Function takes argument (`pd.DataFrame`) that looks like this:

               land  electors  Erya Sterk  Turion Lemmister  Deinuris Targurion
    0    Wunterfull         3         0.6               0.3                 0.1
    1      Ruverren         5         0.3               0.2                 0.5
    2   Vaws Doqwak         2         0.2               0.3                 0.5

In the first column, name of land is given, in the second, number of delegates that corresponds to this land, the other columns are share of the votes that each candidate obtained in that land.

For example, for this table `Wunterfull` winner is `Erya Sterk`, in the other two — `Deinuris Targurion`. Erya will obtain 3 delegate votes and Deinuris — 7 votes. Deinuris will be winner. Function should return `("Deinuris Targurion", 7)`.

If two candidates has equal number of delegate votes, the winner is the first in alphabetical order.

**Note.** No loops and `if`'s are allowed.

**Hint.** You'll probably need `.idxmax()` and `.sort_index()`.

````python
# YOUR CODE HERE

def winner_votes(df):
    
    # choose winner
    df['winner_in_land'] = df[df.columns.values[2:]].idxmax(axis=1)
    
    # get numbers of electors
    df = df[['electors', 'winner_in_land']].groupby('winner_in_land').sum()
    
    # sort
    df.sort_values(['electors', 'winner_in_land'], ascending=[False, True], inplace=True)
    df.reset_index(inplace=True)
    df = df.iloc[0]
    
    return (df['winner_in_land'], df['electors'])
````

````python
# YOUR CODE HERE

def merge_sectors(df_emp, df_names):
    
    # change typ
    df_emp['code2'] = df_emp['code3'].astype("string")
    
    # add column with code2 (delete last digit)
    df_emp['code2'] = df_emp['code2'].str[:-1]
    
    # return int type
    df_emp['code2'] = df_emp['code2'].astype("int")
    
    # merge two dataframes
    res = pd.merge(df_emp, df_names, how="left")

    # rename the column
    res.rename(columns={'name':'group_name'}, inplace=True)
    
    # return dataframe without code2 column
    res.drop('code2', inplace=True, axis=1)
    
    return res
````

For example, for the data shown above, function should return `pd.DataFrame` object with the following contents:

    good     sweeties  chocolate  juice  lemons
    client                                     
    Alice        54.0      18.90    0.0     0.0

**Hint.** Most probably, you need methods `merge`, `fillna`, `pivot_table` and `reindex`. You have to read documentations on these function to discover the correct way to do it.

````python
# YOUR CODE HERE:

def totals(purchases, goods, discounts):
    # rename 
    purchases.rename(columns={'item':'good'}, inplace=True)

    # merg dataframes
    res = pd.merge(goods, purchases, how="left")
    res = pd.merge(res, discounts, how="left")

    # replase NaN
    res = res.fillna(0)

    # calculate Total
    res['total spent'] = res['quantity'] * res['price'] * (1 - res['discount'] / 100)
    
    # drop
    res = res.drop(['quantity', 'price', 'discount'], axis=1)
    
    # build the table
    table = pd.pivot_table(res, values='total spent', index=['client'],
                    columns=['good'], aggfunc=np.sum)
    
    # replace NaN
    table = table.fillna(0)
    
    # remove nulls rows
    # table.dropna(inplace=True) 
    for idx in list(table.index):
        if idx == 0:
            table = table.drop(idx)
    
    return table
````

You have to help professor MacGaggle to find the final grades. They are calculated as an average of all grades student has. If student didn't attend without valid excuse, he/she get 0 for a particular quiz. If the reason of valid, this quiz is simply excluded from counting.

For example, consider this dataset:

````python
import datetime
grades = pd.DataFrame([[5, np.nan, 7, np.nan], 
                       [2, np.nan,      np.nan, 4]], index=['Hannah', 'Ryan'],
                     columns=pd.date_range(start="2017-02-01", freq="W", 
                                           periods=4))
excuses = pd.DataFrame([['Hannah', datetime.datetime(2017, 2, 5), 
                         'was ill'],
                        ['Hannah', datetime.datetime(2017, 2, 12), 
                         'illness'],
                        ['Ryan', datetime.datetime(2017, 2, 19), 'family'],
                        ['Hurray',datetime.datetime(2017, 2, 19), 
                         'sport']],
                      columns=['student', 'date', 'reason'])
````

Hannah didn't attend on 2017-02-12 because she was ill and this work will be excluded. 2017-02-26 she don't have any valid reason so she get 0. Her final grade will be (5 + 7 + 0) / 3 = 4. 

Ryan does not have any valid excuses, so he'll get (2 + 0 + 0 + 4) / 4 = 1.5.

Write function `final_grades(grades, excuses)` that takes two dataframes like in the example above and returns `pd.Series` object, which indexes are student names (in the same order as in `grades`), and values are final grades. For the example above the result is `pd.Series` object of the form:

    Hannah    4.0
    Ryan         1.5
    dtype: float64

**Note.** No loops or `if`'s are allowed. Think pandas.

````python
# YOUR CODE HERE

def final_grades(grades, excuses):
    
    # change df form (date into colums)
    grades = pd.melt(grades, value_vars=grades.columns, ignore_index=False).reset_index()
    
    # move indexes to column (student names) and specify other columns names
    grades.columns = ['student','date', 'value']
    
    # merge data
    merged_data = pd.merge(grades, excuses, on=['student','date'], how='left')
    
    # select data with 'ill' word
    myfilter = lambda x: 'ill' in x
    merged_data['reason'] = merged_data['reason'].fillna('No').apply(myfilter)
    
    # remove NaN scores with exuces = True
    merged_data = merged_data.drop(merged_data[(merged_data['reason'] == True) & (merged_data['value'].isna())].index)
    
    # fill other excuses as 0
    merged_data['value'] = merged_data['value'].fillna(0)
    
    # find average value
    res = merged_data[['student','value']].groupby('student').mean()['value']
    
    return res
````
