```python
import itertools
import time
import csv
from imdb_code import get_movie_distance
```

```python
actors_list = [('Dwayne Johnson', 'https://www.imdb.com/name/nm0425005/'),
               ('Chris Hemsworth', 'https://www.imdb.com/name/nm1165110/'),
               ('Robert Downey Jr.', 'https://www.imdb.com/name/nm0000375/'),
               ('Akshay Kumar', 'https://www.imdb.com/name/nm0474774/'),
               ('Jackie Chan', 'https://www.imdb.com/name/nm0000329/'),
               ('Bradley Cooper', 'https://www.imdb.com/name/nm0177896/'),
               ('Adam Sandler', 'https://www.imdb.com/name/nm0001191/'),
               ('Scarlett Johansson', 'https://www.imdb.com/name/nm0424060/'),
               ('Sofia Vergara', 'https://www.imdb.com/name/nm0005527/'),
               ('Chris Evans', 'https://www.imdb.com/name/nm0262635/')]
```

```python
pair_order_list = list(itertools.combinations(actors_list, 2))
headers = [('actor 1', 'actor 2', 'actor1 -> actor2', 'actor2 -> actor1', 'distance')]
count = 0
file_whole = []
file_ = []
all_time = time.time()

for pair in pair_order_list:
    print(f'\n{count + 1} / {len(pair_order_list)}\n')
    res1 = get_movie_distance(pair[0][1], pair[1][1], 3, 5, 5, logging=False)
    res2 = get_movie_distance(pair[1][1], pair[0][1], 3, 5, 5, logging=False)

    if res1 == res2:
        res = res1
    elif res1 == float('inf') and res2 != float('inf'):
        res = res2
    elif res2 == float('inf') and res1 != float('inf'):
        res = res1
    else:
        res = min(res1, res2)

    file_whole.append((pair[0], pair[1], res1, res2, res))
    file_.append((pair[0][0], pair[1][0], res1, res2, res))
    count += 1

    if count % 9 == 0:
        with open(f'distances_whole{count}.csv', 'w') as f:
            write = csv.writer(f)
            write.writerows(headers)
            write.writerows(file_whole)

with open(f'distances_whole.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(headers)
    write.writerows(file_whole)

with open(f'distances.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(headers)
    write.writerows(file_)

print("--- %s min ---" % ((time.time() - all_time) / 60))
print('All data is collected')
```
